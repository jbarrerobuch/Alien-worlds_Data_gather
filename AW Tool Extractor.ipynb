```python
from daltonapi.api import Atom
import pandas as pd
import datetime
import os.path
#Instanciate dalton api
atom = Atom()
```

# Function definitions

Atomic object to dictionary

```python
def object_to_dict(obj):
    temp_dict={}
    obj_dict = obj.__dict__
    for key,value in obj_dict.items():
        if not type(value) == dict:
            temp_dict[key] = value
        else:
            for subkey,subvalue in value.items():
                temp_dict[subkey] = subvalue
    return temp_dict
```

List of atomic objects to DataFrame

```python
def list_of_objects_to_df(obj_lst):    
    lst_asset_dict = []
    for asset in obj_lst:
        asset_dict = object_to_dict(asset)
        lst_asset_dict.append(asset_dict)
    df = pd.DataFrame(lst_asset_dict)
    return df
```

Save to csv and clean gather DF

```python
def save_and_clean(df,filename):
    if os.path.isfile(filename): #if csv exist append without headers
        df.to_csv(path_or_buf=filename,mode='a',index=False,header=False)
        #print('CSV exists, assets saved\n')
    else: #if csv doesn't exist create file with DF headers
        df.to_csv(path_or_buf=filename,mode='w',index=False)
        #print('CSV created and saved\n')
    df.drop(df.index[:], inplace=True) #wipe rows from DF after saving
    #print('Removed all rows from gather_df\n')
```

# Extraction

```python
collection = 'alien.worlds'
schema = 'tool.worlds'
limit = 1000 #items per page
page = 3250 # Starting page to retrive
asset_count = 0
gathered_count = 0
gather_limit = 50000
output = 'AW_tools_diff.csv'
```

```python
#while True:
while page<3251:
    try:
        Data = atom.get_assets(collection=collection,schema=schema,page=page,limit=limit)
        #print('Get assets succesful')
    except:
        #print('Get assets failed, trying again.\n')
        continue
    else:
        if len(Data) == 0:
            print(f'Cero assets found in page {page}.\n')
            print(f'\nSaving to csv.')
            save_and_clean(gather_df,output)
            break
        else:
            page_df = list_of_objects_to_df(Data)
            #print(f'df created from page {page} with {len(page_df)} assets')
            try:
                gather_df = gather_df.append(page_df, ignore_index= True)
                #print(f'Page appended to gather_df, length is {len(gather_df)}')
            
            except NameError:
                gather_df = list_of_objects_to_df(Data)
                #print(f'gather_df not defined created with page data, {len(gather_df)}')
                
            finally:
                asset_count += len(page_df)
                print(f'Total of {len(gather_df)} assets gathered. Page {page} asset count {asset_count}',end='\r')
                if len(gather_df)>=gather_limit: #reach gathering limit and save to csv
                    print(f'\nGathering limit reached {len(gather_df)}. Saving to csv.\n')
                    save_and_clean(gather_df,output)
                page += 1
    #finally:
        #print(f'iteration ended.')
print('\nExtraction ended\n')
```
